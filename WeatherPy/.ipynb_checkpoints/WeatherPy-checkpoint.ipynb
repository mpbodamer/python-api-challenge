```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
from scipy.stats import linregress

# Import API key
from config import openWeatherKey

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
lat = []
lng = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
        #creating a uniqle list of lats/lng,
        lat.append(lat_lng[0])
        lng.append(lat_lng[1])

#Turning everything into a dictionary

dictOfCities = { i : {"cityName" : f"{i}" ,"temp": 0, "humidity" : 0, "cloudCover" : 0, "windSpeed" : 0,
                      "lat" : 0, "lng" : 0} for i in cities }
listOfCities =list(dictOfCities.values())

for i in range(len(listOfCities)):
    listOfCities[i]["lat"] = lat[i]
    listOfCities[i]["lng"] = lng[i]
    
listOfCities

#Note: There probably is some way to apply lat and lng in the dictOfCities line, but I couldn't figure it out
# The for loop is probably a little brute-y but it works.
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
print("Beginning to get Data")

for i in range(len(listOfCities)):
    cityName = cities[i]
    apiURL = f"http://api.openweathermap.org/data/2.5/weather?q={cities[i]}&units=imperial&APPID={openWeatherKey}"
    print(f"Processing city number {i+1} out of {len(listOfCities)}: {cityName}")
    response = requests.get(apiURL).json()
    
    
    try:
        listOfCities[i]['temp'] = response['main']['temp']
    except KeyError:
        print("KeyError, no Temp")
        listOfCities[i]['temp'] = np.nan
    try:
        listOfCities[i]['humidity'] = response['main']['humidity']
    except KeyError:
        print('KeyError, no humidity')
        listOfCities[i]['humidity'] = np.nan
    try:
        listOfCities[i]['cloudCover'] = response['clouds']['all']
    except KeyError:
        print('KeyError, no Cloudcover')
        listOfCities[i]['cloudCover'] = np.nan
    try:
        listOfCities[i]['windSpeed'] = response['wind']['speed']
    except KeyError:
        print("KeyError, no windspeed")
        listOfCities[i]['windSpeed'] = np.nan
```

```python
cityDataframe = pd.DataFrame(listOfCities)
cityDataframe = cityDataframe.dropna(how = 'any')
cityDataframe
```

### Convert Raw Data to DataFrame

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def linearRegressionPlotter(x,y):
    slope = 0
    intercept = 0
    r_value = 0
    p_value = 0
    std_err = 0
    slope, intercept, r_value, p_value, std_err = linregress(x,y)
    r_squared = r_value * r_value
    line = slope*x+intercept
    
    plt.plot(x, y,'o')
    plt.plot(x, line)
    
    slope = round(slope, 4)
    intercept = round(intercept, 4)
    r_squared = round(r_squared, 4)
    
    label = ("datapoint",f"y = {slope}x + {intercept}")
    plt.legend(label)
    print(f"The r-squared is: {r_squared}")
```

```python
# Create Northern and Southern Hemisphere DataFrames
northernDataframe = cityDataframe[cityDataframe['lat'] >= 0]
southernDataframe = cityDataframe[cityDataframe['lat'] <= 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
plt.title('Northern Hemisphere - Max Temp vs. Latitude Linear Regression')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.grid()
linearRegressionPlotter(northernDataframe['lat'], northernDataframe['temp'])
plt.savefig("output/North_Temp_v_Lat_Linregress")
```
