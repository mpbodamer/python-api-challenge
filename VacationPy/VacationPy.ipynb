## Dependencies

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from pprint import pprint
from citipy import citipy
import gmaps
import os

from apiConfig import googleKey
```

## Store Part I results into DataFrame

```python
country = []
csvPath = "../WeatherPy/cityData.csv"
cityDataframe = pd.read_csv(csvPath)
cityDataframe = cityDataframe.drop(columns=['Unnamed: 0'])

for i in range(len(cityDataframe)):
    city = citipy.nearest_city(cityDataframe["lat"][i],cityDataframe['lng'][i])
    country.append(city.country_code)

cityDataframe['country'] = country
cityDataframe
```

## Humidity Heatmap

```python
gmaps.configure(api_key = googleKey)
fig = gmaps.figure(center = (0,0), zoom_level = 1)
locations = []
for i in range(len(cityDataframe)):
    locations.append((cityDataframe['lat'][i],cityDataframe['lng'][i]))

fig.add_layer(gmaps.heatmap_layer(locations, weights = cityDataframe['humidity'], max_intensity = 200))
fig
```

## Create new DataFrame fitting weather criteria

```python
sortedDataframe = cityDataframe[cityDataframe['temp'] > 70]
sortedDataframe = sortedDataframe[sortedDataframe['temp'] < 80]
sortedDataframe = sortedDataframe[sortedDataframe['windSpeed'] < 10]
sortedDataframe = sortedDataframe[sortedDataframe['cloudCover'] == 0]
sortedDataframe = sortedDataframe[sortedDataframe['humidity'] < 70]

sortedDataframe
```
